```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')

# 한글 기본 폰트 사용 가능 (Mac에서는 다름)
# 만약, 다른 폰트 적용 원하면 
# 참조 : https://dschloe.github.io/python/2023/02/matplotlib_koreanfont/
plt.rcParams['font.family'] = 'Malgun Gothic' # 옵션 설정
fig, ax = plt.subplots(2, 2, figsize=(14, 10))


# 첫번째 그래프
ax0 = sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time", ax=ax[0, 0]) # ax[0, 0] : 위치 조정하기 -> 0번째 행에 0번째 열
ax[0, 0].set_title("총 금액과 팁")

# 두번째 그래프
ax2 = sns.boxplot(data=tips, x="day", y="total_bill", hue="sex", ax=ax[0,1])
ax[0,1].set_title("요일 및 성별에 따른 총 금액")

#세번째 그래프
ax3 = sns.violinplot(data=tips, x="day", y="tip", hue="smoker", ax=ax[1,0])
ax[1,0].set_title("요일 및 흡연 여부에 따른 팁")

#네번째 그래프
ax4 = sns.histplot(data=tips, x="total_bill", bins=20, ax=ax[1,1])
ax[1,1].set_title("총 금액 히스토그램") 
ax[1,1].set_xlabel("총 금액")  # X축의 이름 설정하기
ax[1,1].set_ylabel("빈도")  # Y축의 이름 설정하기 

plt.tight_layout() # 간격 조정
plt.savefig('yourname.png', dpi=300) # 시각화 내보내기
plt.show()
```
